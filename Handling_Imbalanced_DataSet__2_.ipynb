```python
import pandas as pd
import numpy as np
```

```python
Dataset=pd.read_csv('/content/credit_data.csv')
Dataset.tail()
```

```python
#check any missing values
Dataset.isnull().sum()
```

```python
#distribution of data points
Dataset['Class'].value_counts()   #This is Highlt imbalanced data set . 0 - legit transaction , 1 - fraudlant transaction
```

```python
#separating   the legit and fraud
legit=Dataset[Dataset.Class==0]
fraud=Dataset[Dataset.Class==1]

print(legit.shape)
print(fraud.shape)

fraud.tail()
```

**Under Sampling**


Build a sample dataset containing similar distribuition of legit & fraud transactions

```python
legit_sample=legit.sample(n=262)
legit_sample.tail()
legit_sample.shape
```

**Concatenate 2 DataFrames**

```python
n_dataset=pd.concat([legit_sample,fraud],axis=0) #axis 0 means Concatenate one with top of one with row wise
n_dataset.head()  #This is balanced dataset
```
